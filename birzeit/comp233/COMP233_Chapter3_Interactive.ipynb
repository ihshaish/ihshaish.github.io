# Section 3.1 — Predicates and quantified statements I

A **predicate** (*المُسند*) is a sentence containing one or more variables that becomes a statement when specific values are substituted.

Examples:
- `P(x): x > 1/x` is a predicate. `P(2)` says `2 > 1/2`, which is **true**. `P(1/2)` says `1/2 > 2`, which is **false**.
- `Q(n): n is a factor of 8` is a predicate. `Q(4)` is true, `Q(5)` is false.

The **truth set** (*مجموعة الصدق*) of a predicate `P(x)` over a domain `D` is the set `{x ∈ D | P(x)}` — every value in `D` that makes `P(x)` true.

```python
def truth_set(predicate, domain):
    return {x for x in domain if predicate(x)}

domain_options = {
    'Z (integers, –5 to 5)': list(range(-5, 6)),
    'Z+ (positive integers, 1 to 12)': list(range(1, 13)),
    'small set {-2, -1, 0, 1, 2}': [-2, -1, 0, 1, 2],
}

predicate_options = {
    'P(x): x² > x':                lambda x: x*x > x,
    'P(x): x is a factor of 8':    lambda x: x != 0 and 8 % x == 0,
    'P(x): x is even':             lambda x: x % 2 == 0,
    'P(x): x is prime (x ≥ 2)':    lambda x: x >= 2 and all(x % d for d in range(2, int(x**0.5)+1)),
    'P(x): x + 1 > 0':             lambda x: x + 1 > 0,
    'P(x): x² = 1':                lambda x: x*x == 1,
}

dom_dd = widgets.Dropdown(options=list(domain_options.keys()), description='Domain:', style={'description_width': '90px'}, layout=widgets.Layout(width='400px'))
pred_dd = widgets.Dropdown(options=list(predicate_options.keys()), description='Predicate:', style={'description_width': '90px'}, layout=widgets.Layout(width='400px'))
ts_out = widgets.Output()

def show_truth_set(change=None):
    D = domain_options[dom_dd.value]
    pred = predicate_options[pred_dd.value]
    ts = truth_set(pred, D)
    with ts_out:
        ts_out.clear_output()
        print(f"Domain D = {D}")
        print(f"Predicate: {pred_dd.value}")
        print(f"Truth set = {{x ∈ D | P(x)}} = {sorted(ts)}")

dom_dd.observe(show_truth_set, names='value')
pred_dd.observe(show_truth_set, names='value')
display(dom_dd, pred_dd, ts_out)
show_truth_set()
```

## 3.1.1 — The universal quantifier ∀

> ∀x ∈ D, P(x)    reads as    "for all x in D, P(x)"

A universal statement is **true** when P(x) is true for *every* element of D, and **false** when there is even one element for which P(x) is false — a **counterexample** (*المثال المضاد*).

## 3.1.2 — The existential quantifier ∃

> ∃x ∈ D such that P(x)    reads as    "there exists an x in D such that P(x)"

An existential statement is **true** when there is *at least one* element of D making P(x) true — a **witness**. It is **false** only when every element fails.

```python
quant_dd = widgets.Dropdown(options=['∀ (for all)', '∃ (there exists)'], description='Quantifier:', style={'description_width': '90px'}, layout=widgets.Layout(width='400px'))
dom2_dd = widgets.Dropdown(options=list(domain_options.keys()), description='Domain:', style={'description_width': '90px'}, layout=widgets.Layout(width='400px'))
pred2_dd = widgets.Dropdown(options=list(predicate_options.keys()), description='Predicate:', style={'description_width': '90px'}, layout=widgets.Layout(width='400px'))
q_out = widgets.Output()

def evaluate_quant(change=None):
    D = domain_options[dom2_dd.value]
    pred = predicate_options[pred2_dd.value]
    is_universal = quant_dd.value.startswith('∀')
    successes = [x for x in D if pred(x)]
    failures = [x for x in D if not pred(x)]
    with q_out:
        q_out.clear_output()
        print(f"Statement:  {quant_dd.value[0]}x ∈ D, {pred2_dd.value[3:]}")
        print(f"Domain D = {D}")
        print(f"Elements where P(x) is TRUE  : {successes}")
        print(f"Elements where P(x) is FALSE : {failures}")
        print()
        if is_universal:
            if not failures:
                print("✓ Statement is TRUE — P(x) holds for every x in D.")
            else:
                print(f"✗ Statement is FALSE — counterexample: x = {failures[0]} makes P(x) false.")
        else:
            if successes:
                print(f"✓ Statement is TRUE — witness: x = {successes[0]} makes P(x) true.")
            else:
                print("✗ Statement is FALSE — no x in D makes P(x) true.")

quant_dd.observe(evaluate_quant, names='value')
dom2_dd.observe(evaluate_quant, names='value')
pred2_dd.observe(evaluate_quant, names='value')
display(quant_dd, dom2_dd, pred2_dd, q_out)
evaluate_quant()
```

## 3.1.3 — The universal conditional

The most common form of statement in mathematics:

> ∀x ∈ D, P(x) → Q(x)

This is true exactly when every x that satisfies P also satisfies Q. To find counterexamples, we look for an x where P(x) is true but Q(x) is false.

**Equivalent forms** to memorise (Epp §3.1):
- ∀x, P(x) → Q(x)
- "If P(x), then Q(x)" (with x universally quantified)
- "Every P is a Q"
- "P only if Q"
- "P is a sufficient condition for Q"
- "Q is a necessary condition for P"

We will see these in action in Tarski's World below.

## 🌍 Tarski's World

We now build the Tarski World used in **Epp Figure 3.1.1** — a small two-dimensional universe of named objects with shapes and colours, laid out on a grid. Predicates we will use:

| Predicate | Meaning |
|---|---|
| `Triangle(x)` | x is a triangle |
| `Square(x)` | x is a square |
| `Circle(x)` | x is a circle |
| `Blue(x)`, `Gray(x)`, `Black(x)` | x has that colour |
| `RightOf(x, y)` | x is to the right of y |
| `Above(x, y)` | x is above y |
| `SameColor(x, y)` | x and y have the same colour |

Each object also has a name: `a`, `b`, `c`, ..., `k`. Run the next cell to see the world.

```python
# Tarski World from Epp Figure 3.1.1.
# Each object: (name, shape, color, column, row).  Row 1 = bottom, row 4 = top.
# Layout constrained by Epp's worked examples in §3.1.13 and §3.3.10.
TARSKI_311 = [
    ('a', 'circle',   'gray',  1, 4),
    ('b', 'circle',   'blue',  3, 4),
    ('c', 'circle',   'gray',  4, 4),
    ('d', 'triangle', 'blue',  4, 3),
    ('e', 'square',   'blue',  1, 3),
    ('f', 'triangle', 'blue',  1, 2),
    ('g', 'square',   'blue',  3, 2),
    ('h', 'square',   'black', 2, 2),
    ('i', 'triangle', 'blue',  4, 2),
    ('j', 'square',   'blue',  2, 1),
    ('k', 'triangle', 'blue',  4, 1),
]

def make_world(spec):
    world = {}
    for name, shape, color, col, row in spec:
        world[name] = {'name': name, 'shape': shape, 'color': color, 'col': col, 'row': row}
    return world

world = make_world(TARSKI_311)

def draw_world(world, title='Tarski World — Figure 3.1.1', highlight=None, ax=None):
    standalone = ax is None
    if standalone:
        fig, ax = plt.subplots(figsize=(7, 6))
    color_map = {'blue': '#4472B4', 'gray': '#888888', 'black': '#202020'}
    highlight = highlight or set()
    for obj in world.values():
        x, y = obj['col'], obj['row']
        c = color_map[obj['color']]
        edge = '#d4a017' if obj['name'] in highlight else 'white'
        lw = 4 if obj['name'] in highlight else 1.5
        if obj['shape'] == 'triangle':
            patch = RegularPolygon((x, y), numVertices=3, radius=0.32, orientation=0, facecolor=c, edgecolor=edge, linewidth=lw)
        elif obj['shape'] == 'square':
            patch = Rectangle((x-0.28, y-0.28), 0.56, 0.56, facecolor=c, edgecolor=edge, linewidth=lw)
        else:
            patch = MplCircle((x, y), 0.30, facecolor=c, edgecolor=edge, linewidth=lw)
        ax.add_patch(patch)
        text_color = 'white' if obj['color'] in ('blue', 'black') else 'black'
        ax.text(x, y, obj['name'], ha='center', va='center', fontsize=11, fontweight='bold', color=text_color)
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 5)
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title)
    ax.grid(True, alpha=0.2)
    if standalone:
        plt.tight_layout()
        plt.show()

draw_world(world)
```

### Predicate functions

We implement each predicate as a Python function operating on object names from the world.

```python
def Triangle(x):   return world[x]['shape'] == 'triangle'
def Square(x):     return world[x]['shape'] == 'square'
def Circle(x):     return world[x]['shape'] == 'circle'
def Blue(x):       return world[x]['color'] == 'blue'
def Gray(x):       return world[x]['color'] == 'gray'
def Black(x):      return world[x]['color'] == 'black'
def RightOf(x, y): return world[x]['col'] > world[y]['col']
def Above(x, y):   return world[x]['row'] > world[y]['row']
def SameColor(x, y): return world[x]['color'] == world[y]['color']

OBJECTS = list(world.keys())

# Quick sanity check matching Epp Example 3.1.13:
print("Epp 3.1.13 verification (Figure 3.1.1):")
print(f"  ∀t, Triangle(t) → Blue(t)         → {all((not Triangle(t)) or Blue(t) for t in OBJECTS)}  (Epp: TRUE)")
print(f"  ∀x, Blue(x) → Triangle(x)         → {all((not Blue(x)) or Triangle(x) for x in OBJECTS)}  (Epp: FALSE — counterexample e)")
print(f"  ∃y, Square(y) ∧ RightOf(d, y)     → {any(Square(y) and RightOf('d', y) for y in OBJECTS)}  (Epp: TRUE)")
print(f"  ∃z, Square(z) ∧ Gray(z)           → {any(Square(z) and Gray(z) for z in OBJECTS)}  (Epp: FALSE)")
```

All four match Epp's worked answers — the world is correctly set up.

### Single-quantifier statements over Tarski's World

Pick a quantifier and a predicate. The notebook evaluates the statement against the world, highlights the relevant objects, and explains the result.

```python
single_statements = {
    '∀x, Triangle(x) → Blue(x)':         ('forall', lambda x: (not Triangle(x)) or Blue(x), Triangle, 'triangle that is not blue'),
    '∀x, Blue(x) → Triangle(x)':         ('forall', lambda x: (not Blue(x)) or Triangle(x), Blue, 'blue object that is not a triangle'),
    '∀x, Square(x) → Blue(x)':           ('forall', lambda x: (not Square(x)) or Blue(x), Square, 'square that is not blue'),
    '∃x, Square(x) ∧ Black(x)':          ('exists', lambda x: Square(x) and Black(x), None, 'square that is black'),
    '∃x, Square(x) ∧ Gray(x)':           ('exists', lambda x: Square(x) and Gray(x), None, 'square that is gray'),
    '∃x, Circle(x) ∧ Blue(x)':           ('exists', lambda x: Circle(x) and Blue(x), None, 'circle that is blue'),
    '∃x, Triangle(x) ∧ ¬Blue(x)':        ('exists', lambda x: Triangle(x) and not Blue(x), None, 'triangle that is not blue'),
}

ss_dd = widgets.Dropdown(options=list(single_statements.keys()), description='Statement:', style={'description_width': '90px'}, layout=widgets.Layout(width='500px'))
ss_out = widgets.Output()

def eval_single(change=None):
    kind, pred, restrictor, desc = single_statements[ss_dd.value]
    with ss_out:
        ss_out.clear_output()
        print(f"Evaluating: {ss_dd.value}")
        print()
        if kind == 'forall':
            failures = [x for x in OBJECTS if not pred(x)]
            in_domain_of_interest = [x for x in OBJECTS if restrictor(x)] if restrictor else OBJECTS
            print(f"Objects matching the hypothesis: {in_domain_of_interest}")
            if not failures:
                print(f"✓ TRUE — every {desc.split(' that')[0] if 'that' in desc else 'x'} satisfying the hypothesis also satisfies the conclusion.")
            else:
                print(f"✗ FALSE — counterexample: x = {failures[0]} is a {desc}.")
        else:
            successes = [x for x in OBJECTS if pred(x)]
            if successes:
                print(f"✓ TRUE — witness: x = {successes[0]} is a {desc}.")
                print(f"  (All witnesses: {successes})")
            else:
                print(f"✗ FALSE — no {desc} exists in the world.")
        print()
        fig, ax = plt.subplots(figsize=(6, 5))
        highlight = set()
        if kind == 'forall' and failures:
            highlight = {failures[0]}
        elif kind == 'exists':
            highlight = set(x for x in OBJECTS if pred(x))
        draw_world(world, title='Highlighted = relevant objects', highlight=highlight, ax=ax)
        plt.tight_layout()
        plt.show()

ss_dd.observe(eval_single, names='value')
display(ss_dd, ss_out)
eval_single()
```

# Section 3.2 — Predicates and quantified statements II

## 3.2.1 — Negation of quantified statements (النفي)

The two rules to memorise:

$$\neg(\forall x \in D,\, P(x)) \;\equiv\; \exists x \in D \text{ such that } \neg P(x)$$

$$\neg(\exists x \in D \text{ such that } P(x)) \;\equiv\; \forall x \in D,\, \neg P(x)$$

**Pattern:** the negation flips the quantifier and negates the inside.

This is the quantified analogue of De Morgan's laws from Chapter 2. To convince yourself, think:
- "Not everybody is happy"  ≡  "Somebody is unhappy."
- "Nobody is happy" (no one is happy)  ≡  "Everybody is unhappy."

## 3.2.2 — Negation of universal conditionals

$$\neg(\forall x,\, P(x) \to Q(x)) \;\equiv\; \exists x \text{ such that } P(x) \wedge \neg Q(x)$$

**Derivation:**

| Step | Form | Reason |
|---|---|---|
| 1 | ¬(∀x, P(x) → Q(x)) | original |
| 2 | ∃x, ¬(P(x) → Q(x)) | rule for negating ∀ |
| 3 | ∃x, ¬(¬P(x) ∨ Q(x)) | rewrite → as ¬P ∨ Q (from §2.2) |
| 4 | ∃x, P(x) ∧ ¬Q(x) | De Morgan + double negation |

So the negation of *"every P is a Q"* is *"there is a P that is not a Q"* — a **counterexample**.

```python
negation_examples = [
    ('∀x ∈ R, x² ≥ 0',
     '∃x ∈ R, x² < 0',
     'TRUE — every real number squared is non-negative.',
     'FALSE — no such x exists.'),
    ('∀ humans h, mortal(h)',
     '∃ a human h such that ¬mortal(h)',
     'TRUE (assuming the standard biological reading).',
     'FALSE.'),
    ('∃x ∈ Z, x² = 2',
     '∀x ∈ Z, x² ≠ 2',
     'FALSE — no integer squares to 2.',
     'TRUE.'),
    ('∀ basketball players p, tall(p)',
     '∃ a basketball player p such that ¬tall(p)',
     'Depends on the league — but logically: original is FALSE if even one short player exists.',
     'TRUE if such a player exists.'),
    ('∀x, Cat(x) → Mammal(x)',
     '∃x, Cat(x) ∧ ¬Mammal(x)',
     'TRUE — every cat is a mammal.',
     'FALSE.'),
    ('∀x, Bird(x) → CanFly(x)',
     '∃x, Bird(x) ∧ ¬CanFly(x)',
     'FALSE — penguins, ostriches, kiwis are birds that cannot fly.',
     'TRUE — penguins are the witness.'),
]

neg_dd = widgets.Dropdown(
    options=[(ex[0], i) for i, ex in enumerate(negation_examples)],
    description='Statement:',
    style={'description_width': '90px'},
    layout=widgets.Layout(width='600px')
)
neg_out = widgets.Output()

def show_negation(change=None):
    i = neg_dd.value
    orig, neg, orig_val, neg_val = negation_examples[i]
    with neg_out:
        neg_out.clear_output()
        print(f"Original :  {orig}")
        print(f"Negation :  {neg}")
        print()
        print(f"Original is: {orig_val}")
        print(f"Negation is: {neg_val}")
        print()
        print("Notice the negation always has the OPPOSITE truth value of the original.")

neg_dd.observe(show_negation, names='value')
display(neg_dd, neg_out)
show_negation()
```

## 3.2.3 — Vacuous truth (الصدق الفارغ)

A universal conditional ∀x, P(x) → Q(x) is **vacuously true** when **no** x satisfies P(x).

Why? Because the conditional P(x) → Q(x) is automatically true whenever P(x) is false (recall the truth table from §2.2). If no x ever makes the hypothesis true, the conditional cannot fail — so the statement is true by default.

**Example.** "Every elephant in this room is wearing pink shoes." If there are no elephants in the room, the statement is vacuously true.

In Tarski's World below, consider: *∀x, Pentagon(x) → Yellow(x)*. There are no pentagons in our world, so the statement is vacuously true — regardless of any colour condition.

```python
def Pentagon(x):  return False  # No pentagons exist in this Tarski World
def Yellow(x):    return False  # No yellow objects either

# Vacuous-truth demonstration
vacuous_options = {
    '∀x, Pentagon(x) → Yellow(x)':    (lambda x: (not Pentagon(x)) or Yellow(x), Pentagon),
    '∀x, Pentagon(x) → Square(x)':    (lambda x: (not Pentagon(x)) or Square(x), Pentagon),
    '∀x, Pentagon(x) → ¬Square(x)':   (lambda x: (not Pentagon(x)) or not Square(x), Pentagon),
}

vac_dd = widgets.Dropdown(options=list(vacuous_options.keys()), description='Statement:', style={'description_width': '90px'}, layout=widgets.Layout(width='500px'))
vac_out = widgets.Output()

def eval_vacuous(change=None):
    cond, hypo = vacuous_options[vac_dd.value]
    matching = [x for x in OBJECTS if hypo(x)]
    all_hold = all(cond(x) for x in OBJECTS)
    with vac_out:
        vac_out.clear_output()
        print(f"Statement: {vac_dd.value}")
        print(f"Objects satisfying the hypothesis: {matching}")
        print()
        if not matching:
            print("→ No object satisfies the hypothesis.")
            print(f"→ Statement is VACUOUSLY TRUE — the conditional cannot fail.")
        elif all_hold:
            print("→ Statement is TRUE (non-vacuously).")
        else:
            print("→ Statement is FALSE.")

vac_dd.observe(eval_vacuous, names='value')
display(vac_dd, vac_out)
eval_vacuous()
```

## 3.2.4 — Variants of a universal conditional

For the statement ∀x, P(x) → Q(x), the three related forms (parallel to §2.2):

| Form | Symbolic | Equivalent to original? |
|---|---|---|
| Original | ∀x, P(x) → Q(x) | — |
| **Contrapositive** | ∀x, ¬Q(x) → ¬P(x) | **Yes** |
| Converse | ∀x, Q(x) → P(x) | **No** |
| Inverse | ∀x, ¬P(x) → ¬Q(x) | **No** (it is the contrapositive of the converse) |

**Example.** Original: *"Every square is a rectangle."*  (∀x, Square(x) → Rectangle(x))

- Contrapositive: *"Every non-rectangle is a non-square."* ✓ Equivalent and true.
- Converse: *"Every rectangle is a square."* ✗ Not equivalent — false (rectangles can be elongated).
- Inverse: *"Every non-square is a non-rectangle."* ✗ Not equivalent — false (non-square rectangles exist).

**Exercise 3.2.A.** Write a careful negation of each statement.
1. ∀ real numbers x, x > 1/x

# Section 3.3 — Statements with multiple quantifiers

This is where many students stumble. The order of quantifiers matters: **∀∃ is generally NOT the same as ∃∀**.

## 3.3.1 — The challenge-game interpretation (Epp §3.3)

Think of the quantifiers as a two-player game.

**∀x ∈ D, ∃y ∈ E such that P(x, y)** — "for every x, there is a y..."
- Your *opponent* picks any x they want from D.
- *You* respond with some y from E.
- You win if P(x, y) holds. You're allowed to pick a different y for each x.
- The statement is TRUE if you have a winning strategy.

**∃x ∈ D such that ∀y ∈ E, P(x, y)** — "there is an x such that for every y..."
- *You* commit to one specific x first.
- Then your opponent picks any y they want.
- You win if P(x, y) holds. Your x is fixed — it must work against every possible y.
- The statement is TRUE if you can find such a one-size-fits-all x.

The order of moves is what makes the two statements different.

## 3.3.2 — Tarski World — Figure 3.3.1

For §3.3, Epp uses a slightly different layout — **Figure 3.3.1** — that supports the worked examples about quantifier order. We build it below.

```python
# Tarski World from Epp Figure 3.3.1, constrained by Epp's worked examples §3.3.1, 3.3.2, 3.3.8, 3.3.9.
# Established facts: triangles d, f, i (d black; f, i gray); squares e, g, h, j (e black; g, h gray; j blue);
# circles a, b, c. d and i are to the right of all circles; no triangle is to the right of square g or j.
TARSKI_331 = [
    ('a', 'circle',   'blue', 1, 4),
    ('b', 'circle',   'blue', 2, 4),
    ('c', 'circle',   'blue', 3, 4),
    ('d', 'triangle', 'black', 4, 3),
    ('e', 'square',   'black', 1, 3),
    ('f', 'triangle', 'gray',  2, 2),
    ('g', 'square',   'gray',  1, 2),
    ('h', 'square',   'gray',  3, 2),
    ('i', 'triangle', 'gray',  4, 2),
    ('j', 'square',   'blue',  2, 1),
]

world331 = make_world(TARSKI_331)

# Re-bind predicates to operate on the new world
def T(x): return world331[x]['shape'] == 'triangle'
def Sq(x): return world331[x]['shape'] == 'square'
def Ci(x): return world331[x]['shape'] == 'circle'
def Bl(x): return world331[x]['color'] == 'blue'
def Gr(x): return world331[x]['color'] == 'gray'
def Bk(x): return world331[x]['color'] == 'black'
def RO(x, y): return world331[x]['col'] > world331[y]['col']
def Ab(x, y): return world331[x]['row'] > world331[y]['row']
def SC(x, y): return world331[x]['color'] == world331[y]['color']

OBJS = list(world331.keys())

draw_world(world331, title='Tarski World — Figure 3.3.1')

# Verify against Epp's stated facts (§3.3.1, §3.3.2)
print("Verifying against Epp §3.3.1: 'for all triangles x, there is a square y of the same color':")
result = all((not T(x)) or any(Sq(y) and SC(x, y) for y in OBJS) for x in OBJS)
print(f"  {result}  (Epp: TRUE)")
print()
print("Verifying against Epp §3.3.2: 'there is a triangle x such that for all circles y, x is to the right of y':")
result = any(T(x) and all((not Ci(y)) or RO(x, y) for y in OBJS) for x in OBJS)
print(f"  {result}  (Epp: TRUE — witnesses are d and i)")
```

## 3.3.3 — ∀∃ vs ∃∀ side-by-side

Pick a `(shape₁, shape₂, relation)` combination below. The notebook evaluates both quantifier orderings and walks through the challenge-game reasoning.

```python
multi_examples = [
    {
        'label':  '(a) ∀ triangles x, ∃ a square y such that SameColor(x, y)   /   ∃ a square y such that ∀ triangles x, SameColor(x, y)',
        'shape1': T, 'name1': 'triangle',
        'shape2': Sq, 'name2': 'square',
        'rel':    SC, 'rel_name': 'SameColor',
    },
    {
        'label':  '(b) ∀ squares x, ∃ a triangle y such that ¬SameColor(x, y)   /   ∃ a triangle y such that ∀ squares x, ¬SameColor(x, y)',
        'shape1': Sq, 'name1': 'square',
        'shape2': T,  'name2': 'triangle',
        'rel':    lambda x, y: not SC(x, y), 'rel_name': '¬SameColor',
    },
    {
        'label':  '(c) ∀ circles x, ∃ a square y such that RightOf(y, x)   /   ∃ a square y such that ∀ circles x, RightOf(y, x)',
        'shape1': Ci, 'name1': 'circle',
        'shape2': Sq, 'name2': 'square',
        'rel':    lambda x, y: RO(y, x), 'rel_name': 'RightOf(y, x)',
    },
    {
        'label':  '(d) ∀ triangles x, ∃ a circle y such that RightOf(x, y)   /   ∃ a circle y such that ∀ triangles x, RightOf(x, y)',
        'shape1': T, 'name1': 'triangle',
        'shape2': Ci, 'name2': 'circle',
        'rel':    lambda x, y: RO(x, y), 'rel_name': 'RightOf(x, y)',
    },
]

mq_dd = widgets.Dropdown(
    options=[(ex['label'], i) for i, ex in enumerate(multi_examples)],
    description='Example:',
    style={'description_width': '70px'},
    layout=widgets.Layout(width='850px')
)
mq_out = widgets.Output()

def eval_multi(change=None):
    ex = multi_examples[mq_dd.value]
    S1, S2 = ex['shape1'], ex['shape2']
    R = ex['rel']
    n1, n2, rn = ex['name1'], ex['name2'], ex['rel_name']
    domain1 = [x for x in OBJS if S1(x)]
    domain2 = [y for y in OBJS if S2(y)]
    with mq_out:
        mq_out.clear_output()
        # ∀∃ direction
        print(f"=== ∀ {n1}s x, ∃ a {n2} y such that {rn}(x, y) ===")
        print(f"{n1}s in world: {domain1}")
        print(f"{n2}s in world: {domain2}")
        print()
        all_ok = True
        for x in domain1:
            wits = [y for y in domain2 if R(x, y)]
            if wits:
                print(f"  For x = {x}: a working y is {wits[0]}  (all: {wits})")
            else:
                all_ok = False
                print(f"  For x = {x}: NO {n2} y works.  ← breaks ∀∃")
        print(f"\n→ ∀∃ statement is {'TRUE' if all_ok else 'FALSE'}.\n")
        # ∃∀ direction
        print(f"=== ∃ a {n2} y such that ∀ {n1}s x, {rn}(x, y) ===")
        print(f"We need ONE specific y that works for every x.")
        winning_ys = []
        for y in domain2:
            if all(R(x, y) for x in domain1):
                winning_ys.append(y)
        if winning_ys:
            print(f"  Witnesses (y values that work for all x): {winning_ys}")
            print(f"→ ∃∀ statement is TRUE.")
        else:
            print(f"  No single y works for every x. Try each:")
            for y in domain2:
                fails = [x for x in domain1 if not R(x, y)]
                if fails:
                    print(f"    y = {y}: fails for x = {fails[0]}")
            print(f"→ ∃∀ statement is FALSE.")
        print()
        print(f"Compare: ∀∃ = {'TRUE' if all_ok else 'FALSE'} ;   ∃∀ = {'TRUE' if winning_ys else 'FALSE'}")
        if (all_ok) != (bool(winning_ys)):
            print(f"→ The two statements have OPPOSITE truth values — order of quantifiers matters!")
        else:
            print(f"→ In this case the two orderings happen to agree, but in general they need not.")

mq_dd.observe(eval_multi, names='value')
display(mq_dd, mq_out)
eval_multi()
```

## 3.3.4 — Formal logical notation (Epp §3.3.10)

In computer science, statements are often written in fully formal first-order logic:

- "∀x ∈ D, P(x)"   →   ∀x (x ∈ D → P(x))
- "∃x ∈ D such that P(x)"   →   ∃x (x ∈ D ∧ P(x))

For Tarski's World where the domain is implicit, this becomes:

- **"For all circles x, x is above f"**   →   ∀x (Circle(x) → Above(x, f))
- **"There is a square x such that x is black"**   →   ∃x (Square(x) ∧ Black(x))
- **"For all circles x, there is a square y such that x and y have the same colour"**   →   ∀x (Circle(x) → ∃y (Square(y) ∧ SameColor(x, y)))
- **"There is a square x such that for all triangles y, x is to the right of y"**   →   ∃x (Square(x) ∧ ∀y (Triangle(y) → RightOf(x, y)))

**Key pattern:**
- Universal statements use **→** to attach the restriction (it's an *if-then*).
- Existential statements use **∧** to attach the restriction (it's an *and*).

This asymmetry comes from how the quantifier interacts with truth values: a universal statement should be vacuously true outside its restriction, and an existential statement should fail outside its restriction.

## 3.3.5 — Negating multiply-quantified statements

The rule: **walk through the quantifiers left to right, flipping each as you go, and negate the innermost predicate.**

$$\neg\big(\forall x \in D,\, \exists y \in E,\, P(x, y)\big) \;\equiv\; \exists x \in D,\, \forall y \in E,\, \neg P(x, y)$$

$$\neg\big(\exists x \in D,\, \forall y \in E,\, P(x, y)\big) \;\equiv\; \forall x \in D,\, \exists y \in E,\, \neg P(x, y)$$

```python
neg_multi = [
    ('∀x, ∃y such that x + y = 0',
     '∃x such that ∀y, x + y ≠ 0',
     'Flip ∀ to ∃, flip ∃ to ∀, negate the inside.'),
    ('∃x such that ∀y, x · y = y',
     '∀x, ∃y such that x · y ≠ y',
     'Flip ∃ to ∀, flip ∀ to ∃, negate the inside.'),
    ('∀ triangles x, ∃ a square y such that SameColor(x, y)',
     '∃ a triangle x such that ∀ squares y, ¬SameColor(x, y)',
     'Each quantifier flips; SameColor becomes ¬SameColor.'),
    ('∃ a person p such that ∀ people q, loves(p, q)',
     '∀ people p, ∃ a person q such that ¬loves(p, q)',
     '"There is somebody loved by all" negates to "for everybody there is someone they do not love".'),
]

nm_dd = widgets.Dropdown(
    options=[(ex[0], i) for i, ex in enumerate(neg_multi)],
    description='Statement:',
    style={'description_width': '90px'},
    layout=widgets.Layout(width='700px')
)
nm_out = widgets.Output()

def show_nm(change=None):
    i = nm_dd.value
    orig, neg, why = neg_multi[i]
    with nm_out:
        nm_out.clear_output()
        print(f"Original:  {orig}")
        print(f"Negation:  {neg}")
        print()
        print(f"How:  {why}")

nm_dd.observe(show_nm, names='value')
display(nm_dd, nm_out)
show_nm()
```
