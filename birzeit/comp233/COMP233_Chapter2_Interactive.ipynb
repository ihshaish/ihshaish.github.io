# Section 2.1 — Logical form and logical equivalence

A **statement** (or proposition — *العبارة / القضية*) is a sentence that is either true or false, but not both.

Examples of statements:
- "2 + 2 = 4"  → true
- "It is raining in Nablus right now"  → either true or false
- "The number 17 is prime"  → true

Not statements:
- "What time is it?" (a question)
- "Close the door." (a command)
- "x + 1 = 5" (depends on x — it's a *predicate*, which we'll meet in Chapter 3)

In propositional logic we label statements with letters like `p`, `q`, `r` and combine them using the connectives below.

## 2.1.1 — The basic connectives

| Symbol | Name | Read as | Python |
|---|---|---|---|
| ¬ p or ~p | Negation | "not p" | `not p` |
| p ∧ q | Conjunction | "p and q" | `p and q` |
| p ∨ q | Disjunction | "p or q" (inclusive) | `p or q` |

Toggle the two checkboxes below to see how the connectives evaluate.

```python
p_box = widgets.Checkbox(value=True, description='p')
q_box = widgets.Checkbox(value=True, description='q')
out = widgets.Output()

def show_connectives(change=None):
    p = p_box.value
    q = q_box.value
    with out:
        out.clear_output()
        print(f"  p       = {p}")
        print(f"  q       = {q}")
        print(f"  ¬p      = {not p}")
        print(f"  ¬q      = {not q}")
        print(f"  p ∧ q   = {p and q}")
        print(f"  p ∨ q   = {p or q}")

p_box.observe(show_connectives, names='value')
q_box.observe(show_connectives, names='value')
display(widgets.HBox([p_box, q_box]), out)
show_connectives()
```

## 2.1.2 — Truth tables

A **truth table** lists every possible combination of truth values for the inputs and shows the result for each row. With *n* variables there are 2^n rows.

The cell below builds a truth table for any of the standard connectives.

```python
expr_dropdown = widgets.Dropdown(
    options=[
        ('p ∧ q', 'p and q'),
        ('p ∨ q', 'p or q'),
        ('¬p', 'not p'),
        ('¬(p ∧ q)', 'not (p and q)'),
        ('¬p ∨ ¬q', '(not p) or (not q)'),
        ('p ∧ (q ∨ r)', 'p and (q or r)'),
        ('(p ∧ q) ∨ (p ∧ r)', '(p and q) or (p and r)'),
    ],
    description='Expression:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='400px')
)

tt_out = widgets.Output()

def build_truth_table(change=None):
    expr_py = expr_dropdown.value
    expr_label = expr_dropdown.label
    var_names = [v for v in ['p', 'q', 'r'] if v in expr_py]
    n = len(var_names)
    with tt_out:
        tt_out.clear_output()
        header = " | ".join(var_names) + f" | {expr_label}"
        print(header)
        print("-" * len(header))
        for values in product([False, True], repeat=n):
            env = dict(zip(var_names, values))
            result = eval(expr_py, {"__builtins__": {}}, env)
            row = " | ".join("T" if env[v] else "F" for v in var_names)
            print(f"{row} | {'T' if result else 'F'}")

expr_dropdown.observe(build_truth_table, names='value')
display(expr_dropdown, tt_out)
build_truth_table()
```

**Observation.** Notice that the last two expressions in the dropdown, `p ∧ (q ∨ r)` and `(p ∧ q) ∨ (p ∧ r)`, give identical columns. They are **logically equivalent** — this is the *distributive law*. We explore equivalence next.

## 2.1.3 — Logical equivalence (التكافؤ المنطقي)

Two statement forms are **logically equivalent** if they have the same truth value in every row of the truth table. We write `P ≡ Q`.

Pick two expressions below. The notebook builds both truth tables and reports whether they are equivalent.

```python
expr_options = [
    ('p ∧ q', 'p and q'),
    ('q ∧ p', 'q and p'),
    ('p ∨ q', 'p or q'),
    ('¬(p ∧ q)', 'not (p and q)'),
    ('¬p ∨ ¬q', '(not p) or (not q)'),
    ('¬(p ∨ q)', 'not (p or q)'),
    ('¬p ∧ ¬q', '(not p) and (not q)'),
    ('p ∧ (q ∨ r)', 'p and (q or r)'),
    ('(p ∧ q) ∨ (p ∧ r)', '(p and q) or (p and r)'),
    ('p ∨ (q ∧ r)', 'p or (q and r)'),
    ('(p ∨ q) ∧ (p ∨ r)', '(p or q) and (p or r)'),
    ('p', 'p'),
    ('p ∧ p', 'p and p'),
    ('p ∨ p', 'p or p'),
]

left = widgets.Dropdown(options=expr_options, value='p and (q or r)', description='Left:', layout=widgets.Layout(width='350px'))
right = widgets.Dropdown(options=expr_options, value='(p and q) or (p and r)', description='Right:', layout=widgets.Layout(width='350px'))
eq_out = widgets.Output()

def check_equiv(change=None):
    lp, rp = left.value, right.value
    ll, rl = left.label, right.label
    var_names = sorted(set(v for v in ['p', 'q', 'r'] if v in lp or v in rp))
    n = len(var_names)
    with eq_out:
        eq_out.clear_output()
        header = " | ".join(var_names) + f" |  {ll}  |  {rl}"
        print(header)
        print("-" * len(header))
        all_equal = True
        for values in product([False, True], repeat=n):
            env = dict(zip(var_names, values))
            l_val = eval(lp, {"__builtins__": {}}, env)
            r_val = eval(rp, {"__builtins__": {}}, env)
            if l_val != r_val:
                all_equal = False
            row = " | ".join("T" if env[v] else "F" for v in var_names)
            mark = "  ←  differ" if l_val != r_val else ""
            print(f"{row} |  {'T' if l_val else 'F'}  |  {'T' if r_val else 'F'}{mark}")
        print()
        if all_equal:
            print(f"✓ Equivalent: {ll}  ≡  {rl}")
        else:
            print(f"✗ Not equivalent.")

left.observe(check_equiv, names='value')
right.observe(check_equiv, names='value')
display(left, right, eq_out)
check_equiv()
```

## 2.1.4 — De Morgan's laws (قوانين دي مورغان)

Two of the most useful equivalences in this course. They appear constantly in proofs, set theory, programming, and circuit design.

$$\neg(p \wedge q) \;\equiv\; \neg p \vee \neg q$$

$$\neg(p \vee q) \;\equiv\; \neg p \wedge \neg q$$

**Memory aid:** *"Break the bar, change the sign."* When ¬ moves inside the parentheses, ∧ flips to ∨ (and vice versa), and each part gets negated.

Verify them with the equivalence checker above — try `¬(p ∧ q)` vs `¬p ∨ ¬q`, then `¬(p ∨ q)` vs `¬p ∧ ¬q`.

Below is a **practical application**: filtering student records.

```python
students = [
    {"name": "Ali",    "grade": 85, "attendance": 90},
    {"name": "Sara",   "grade": 55, "attendance": 80},
    {"name": "Omar",   "grade": 72, "attendance": 60},
    {"name": "Lina",   "grade": 91, "attendance": 95},
    {"name": "Khaled", "grade": 60, "attendance": 75},
    {"name": "Nour",   "grade": 45, "attendance": 40},
]

grade_threshold = widgets.IntSlider(value=60, min=0, max=100, step=5, description='Pass grade:', style={'description_width': '120px'})
att_threshold = widgets.IntSlider(value=75, min=0, max=100, step=5, description='Min attendance:', style={'description_width': '120px'})
filter_out = widgets.Output()

def filter_students(change=None):
    g_t = grade_threshold.value
    a_t = att_threshold.value
    with filter_out:
        filter_out.clear_output()
        print(f"Rule: qualifies  ≡  (grade ≥ {g_t})  ∧  (attendance ≥ {a_t})")
        print(f"De Morgan: ¬qualifies  ≡  (grade < {g_t})  ∨  (attendance < {a_t})")
        print()
        print(f"{'Name':<8} {'grade':>6} {'att':>5} {'passes':>8} {'att_ok':>8} {'qualifies':>11} {'¬qualifies (DM)':>18}")
        for s in students:
            passed = s['grade'] >= g_t
            att_ok = s['attendance'] >= a_t
            qual = passed and att_ok
            not_qual_dm = (not passed) or (not att_ok)
            print(f"{s['name']:<8} {s['grade']:>6} {s['attendance']:>5} {str(passed):>8} {str(att_ok):>8} {str(qual):>11} {str(not_qual_dm):>18}")
        print()
        print("Notice: the 'qualifies' column and the negation of the '¬qualifies' column always agree.")

grade_threshold.observe(filter_students, names='value')
att_threshold.observe(filter_students, names='value')
display(grade_threshold, att_threshold, filter_out)
filter_students()
```

## 2.1.5 — Tautology and contradiction

- A **tautology** (*التطابق*) is a statement form that is true in **every** row of its truth table. Example: `p ∨ ¬p`.
- A **contradiction** (*التناقض*) is a statement form that is false in **every** row. Example: `p ∧ ¬p`.
- Anything else is a **contingency** — sometimes true, sometimes false.

The cell below classifies a chosen expression.

```python
class_options = [
    ('p ∨ ¬p', 'p or (not p)'),
    ('p ∧ ¬p', 'p and (not p)'),
    ('p ∨ q', 'p or q'),
    ('(p ∧ q) ∨ (¬p ∨ ¬q)', '(p and q) or ((not p) or (not q))'),
    ('p ∧ (¬p ∨ q)', 'p and ((not p) or q)'),
    ('¬(p ∧ ¬p)', 'not (p and (not p))'),
]

class_dropdown = widgets.Dropdown(options=class_options, description='Expression:', style={'description_width': '120px'}, layout=widgets.Layout(width='400px'))
class_out = widgets.Output()

def classify(change=None):
    expr_py = class_dropdown.value
    label = class_dropdown.label
    var_names = sorted(set(v for v in ['p', 'q', 'r'] if v in expr_py))
    results = []
    for values in product([False, True], repeat=len(var_names)):
        env = dict(zip(var_names, values))
        results.append(eval(expr_py, {"__builtins__": {}}, env))
    with class_out:
        class_out.clear_output()
        print(f"Expression: {label}")
        print(f"Truth values across all rows: {results}")
        if all(results):
            print("→ Tautology (التطابق) — always true.")
        elif not any(results):
            print("→ Contradiction (التناقض) — always false.")
        else:
            print("→ Contingency — sometimes true, sometimes false.")

class_dropdown.observe(classify, names='value')
display(class_dropdown, class_out)
classify()
```

# Section 2.2 — Conditional statements

The conditional `p → q` (read "if p then q") is the workhorse of mathematical reasoning. It is defined by the following truth table:

| p | q | p → q |
|---|---|-------|
| T | T | T |
| T | F | **F** |
| F | T | T |
| F | F | T |

The only way `p → q` is false is when p is true but q is false. The rows where p is false make the conditional **vacuously true** — a topic we revisit in 3.2.

```python
p_b = widgets.Checkbox(value=True, description='p')
q_b = widgets.Checkbox(value=True, description='q')
cond_out = widgets.Output()

def show_cond(change=None):
    p = p_b.value
    q = q_b.value
    cond = (not p) or q
    with cond_out:
        cond_out.clear_output()
        print(f"  p = {p}")
        print(f"  q = {q}")
        print(f"  p → q  =  {cond}")
        if not p:
            print("  (vacuously true — hypothesis is false)")

p_b.observe(show_cond, names='value')
q_b.observe(show_cond, names='value')
display(widgets.HBox([p_b, q_b]), cond_out)
show_cond()
```

## 2.2.1 — Contrapositive, converse, inverse

Given the conditional `p → q`, three related forms come up constantly:

| Form | Symbolic | Equivalent to original? |
|---|---|---|
| Original | p → q | — |
| **Contrapositive** | ¬q → ¬p | **Yes** — always equivalent |
| Converse | q → p | **No** — not equivalent |
| Inverse | ¬p → ¬q | **No** — not equivalent (but it's the contrapositive of the converse) |

The biggest student confusion in this chapter is mixing up the contrapositive with the converse. The truth table below settles it for any input.

```python
cp_out = widgets.Output()

def show_variants():
    with cp_out:
        cp_out.clear_output()
        header = " p | q || p→q | ¬q→¬p (contra) | q→p (conv) | ¬p→¬q (inv)"
        print(header)
        print("-" * len(header))
        for p, q in product([False, True], repeat=2):
            orig = (not p) or q
            contra = (not (not q)) or (not p)
            conv = (not q) or p
            inv = (not (not p)) or (not q)
            pp = 'T' if p else 'F'
            qq = 'T' if q else 'F'
            print(f" {pp} | {qq} ||  {'T' if orig else 'F'}  |       {'T' if contra else 'F'}        |     {'T' if conv else 'F'}      |     {'T' if inv else 'F'}")
        print()
        print("Columns to compare:")
        print("  p→q   vs   ¬q→¬p   →   identical in every row → EQUIVALENT (contrapositive).")
        print("  p→q   vs   q→p     →   differ in some row     → NOT equivalent (converse).")
        print("  p→q   vs   ¬p→¬q   →   differ in some row     → NOT equivalent (inverse).")
        print("  Also note:  q→p  ≡  ¬p→¬q  (converse and inverse ARE equivalent to each other).")

display(cp_out)
show_variants()
```

**Worked example (Epp 2.2).** *"If it is raining, then the ground is wet."*

- Original: rain → wet
- Contrapositive: ¬wet → ¬rain — *"If the ground is not wet, then it is not raining."* ✓ Logically the same.
- Converse: wet → rain — *"If the ground is wet, then it is raining."* ✗ Could be wet from a sprinkler.
- Inverse: ¬rain → ¬wet — *"If it is not raining, then the ground is not wet."* ✗ Same flaw as converse.

## 2.2.2 — Biconditional

The **biconditional** `p ↔ q` (read "p if and only if q", abbreviated *iff*) is true exactly when p and q have the same truth value.

A central equivalence:
$$p \leftrightarrow q \;\equiv\; (p \to q) \wedge (q \to p)$$

Verify it below.

```python
bicond_out = widgets.Output()

def show_bicond():
    with bicond_out:
        bicond_out.clear_output()
        header = " p | q || p↔q | (p→q) ∧ (q→p)"
        print(header)
        print("-" * len(header))
        match = True
        for p, q in product([False, True], repeat=2):
            bic = (p == q)
            both = ((not p) or q) and ((not q) or p)
            if bic != both:
                match = False
            pp = 'T' if p else 'F'
            qq = 'T' if q else 'F'
            print(f" {pp} | {qq} ||  {'T' if bic else 'F'}  |      {'T' if both else 'F'}")
        print()
        print("✓ Equivalent." if match else "✗ Differ.")

display(bicond_out)
show_bicond()
```

## 2.2.3 — Necessary and sufficient conditions

These two phrases are crucial, and easy to swap by mistake.

- "*p is **sufficient** for q*"  means  **p → q**.  (Having p is enough to guarantee q.)
- "*p is **necessary** for q*"  means  **q → p**.  (Without p, q cannot happen — so if q, then p.)
- "*p **only if** q*"  means  **p → q**.  (p can only be true when q is.)
- "*p **if** q*"  means  **q → p**.

```python
nec_suff_options = [
    ('Being a square is sufficient for being a rectangle.',
     'square → rectangle',
     'If a shape is a square, then it is a rectangle. (True — squares are special rectangles.)'),
    ('Having a valid ID is necessary for entering the exam hall.',
     'enter → has_valid_ID',
     'You can enter the exam hall only if you have a valid ID. Equivalently: if you enter, you have an ID.'),
    ('You pass only if you score at least 60.',
     'pass → score ≥ 60',
     'Passing requires scoring at least 60. Note: scoring 60+ alone does not guarantee passing — attendance might matter too.'),
    ('A number is divisible by 4 if it is divisible by 8.',
     'divisible_by_8 → divisible_by_4',
     '"if" introduces the hypothesis. So: divisible by 8 implies divisible by 4. (True.)'),
]

ns_dropdown = widgets.Dropdown(
    options=[(s[0], i) for i, s in enumerate(nec_suff_options)],
    description='Statement:',
    style={'description_width': '90px'},
    layout=widgets.Layout(width='600px')
)
ns_out = widgets.Output()

def show_ns(change=None):
    i = ns_dropdown.value
    statement, symbolic, comment = nec_suff_options[i]
    with ns_out:
        ns_out.clear_output()
        print(f"Statement:  {statement}")
        print(f"Symbolic :  {symbolic}")
        print(f"Note     :  {comment}")

ns_dropdown.observe(show_ns, names='value')
display(ns_dropdown, ns_out)
show_ns()
```

**Exercise 2.2.C.** Use the equivalence checker (in Section 2.1) or a hand-truth-table to verify:
$$p \to q \;\equiv\; \neg p \vee q$$

Why is this useful?

```python
sol_2c = widgets.Output()
btn_2c = widgets.Button(description='Show solution', button_style='info')

def show_2c(b):
    with sol_2c:
        sol_2c.clear_output()
        print("Truth table:")
        print()
        print(" p | q || p→q | ¬p ∨ q")
        print("-----------------------")
        for p, q in product([False, True], repeat=2):
            cond = (not p) or q
            disj = (not p) or q
            pp = 'T' if p else 'F'
            qq = 'T' if q else 'F'
            print(f" {pp} | {qq} ||  {'T' if cond else 'F'}  |   {'T' if disj else 'F'}")
        print()
        print("Identical in every row → equivalent.")
        print()
        print("Why it matters:")
        print("  This is THE bridge from conditional reasoning to the basic connectives.")
        print("  It lets us apply De Morgan to negate conditionals:")
        print("      ¬(p → q)  ≡  ¬(¬p ∨ q)  ≡  p ∧ ¬q")
        print("  So the negation of 'if p then q' is 'p AND not q' — used constantly in proofs.")

btn_2c.on_click(show_2c)
display(btn_2c, sol_2c)
```

# Section 2.3 — Valid arguments and rules of inference

An **argument** is a sequence of statements: some **premises** followed by a **conclusion**.

An argument is **valid** (*الحجة الصحيحة*) when:
> in **every** row of the truth table where all premises are true, the conclusion is also true.

If there is even one row where every premise is true but the conclusion is false, the argument is **invalid** (an *invalid* row is called a counterexample to the argument).

Important: validity is about *form*, not about whether the premises are actually true. An argument can be valid with false premises and an absurd conclusion — what matters is that the form preserves truth.

## 2.3.1 — The argument validity checker

Pick an argument form below. The notebook builds the truth table and reports validity, highlighting the *critical rows* (rows where every premise is true).

```python
arguments = {
    'Modus ponens   (p → q, p ⊢ q)': {
        'premises': ['(not p) or q', 'p'],
        'premise_labels': ['p → q', 'p'],
        'conclusion': 'q',
        'conclusion_label': 'q',
    },
    'Modus tollens  (p → q, ¬q ⊢ ¬p)': {
        'premises': ['(not p) or q', 'not q'],
        'premise_labels': ['p → q', '¬q'],
        'conclusion': 'not p',
        'conclusion_label': '¬p',
    },
    'Disjunctive syllogism  (p ∨ q, ¬p ⊢ q)': {
        'premises': ['p or q', 'not p'],
        'premise_labels': ['p ∨ q', '¬p'],
        'conclusion': 'q',
        'conclusion_label': 'q',
    },
    'Hypothetical syllogism  (p → q, q → r ⊢ p → r)': {
        'premises': ['(not p) or q', '(not q) or r'],
        'premise_labels': ['p → q', 'q → r'],
        'conclusion': '(not p) or r',
        'conclusion_label': 'p → r',
    },
    'Converse error (INVALID)  (p → q, q ⊢ p)': {
        'premises': ['(not p) or q', 'q'],
        'premise_labels': ['p → q', 'q'],
        'conclusion': 'p',
        'conclusion_label': 'p',
    },
    'Inverse error (INVALID)  (p → q, ¬p ⊢ ¬q)': {
        'premises': ['(not p) or q', 'not p'],
        'premise_labels': ['p → q', '¬p'],
        'conclusion': 'not q',
        'conclusion_label': '¬q',
    },
}

arg_dropdown = widgets.Dropdown(options=list(arguments.keys()), description='Argument:', style={'description_width': '100px'}, layout=widgets.Layout(width='550px'))
arg_out = widgets.Output()

def check_argument(change=None):
    arg = arguments[arg_dropdown.value]
    premises = arg['premises']
    p_labels = arg['premise_labels']
    conclusion = arg['conclusion']
    c_label = arg['conclusion_label']
    var_names = sorted(set(v for v in ['p', 'q', 'r'] if any(v in p for p in premises) or v in conclusion))
    with arg_out:
        arg_out.clear_output()
        pcols = "  ".join(f"({pl})" for pl in p_labels)
        header = " | ".join(var_names) + f" | {pcols} | ({c_label})"
        print(header)
        print("-" * len(header))
        valid = True
        any_critical = False
        for values in product([False, True], repeat=len(var_names)):
            env = dict(zip(var_names, values))
            p_vals = [eval(pr, {"__builtins__": {}}, env) for pr in premises]
            c_val = eval(conclusion, {"__builtins__": {}}, env)
            all_premises_true = all(p_vals)
            critical = all_premises_true
            if critical:
                any_critical = True
                if not c_val:
                    valid = False
            row = " | ".join("T" if env[v] else "F" for v in var_names)
            pstr = "  ".join(f"  {'T' if v else 'F'}  " for v in p_vals)
            cstr = f"  {'T' if c_val else 'F'}"
            mark = "  ← critical row" if critical else ""
            if critical and not c_val:
                mark += " (BREAKS validity)"
            print(f"{row} | {pstr} | {cstr}{mark}")
        print()
        if not any_critical:
            print("⚠ No row makes all premises true — argument is *vacuously* valid (true premises never occur in this form).")
        elif valid:
            print(f"✓ VALID. In every row where all premises are true, the conclusion is also true.")
        else:
            print(f"✗ INVALID. There is a row where all premises are true but the conclusion is false.")

arg_dropdown.observe(check_argument, names='value')
display(arg_dropdown, arg_out)
check_argument()
```

## 2.3.2 — Key valid forms to memorise

| Name | Form | Reads as |
|---|---|---|
| Modus ponens | p → q, p ⊢ q | "If p then q. p. Therefore q." |
| Modus tollens | p → q, ¬q ⊢ ¬p | "If p then q. Not q. Therefore not p." |
| Disjunctive syllogism | p ∨ q, ¬p ⊢ q | "p or q. Not p. Therefore q." |
| Hypothetical syllogism | p → q, q → r ⊢ p → r | "If p then q. If q then r. Therefore if p then r." |
| Conjunction | p, q ⊢ p ∧ q | — |
| Simplification | p ∧ q ⊢ p | — |

And the two **classic fallacies** to avoid:

| Name | Form | Why invalid |
|---|---|---|
| Converse error | p → q, q ⊢ p | Confuses the conditional with its converse |
| Inverse error | p → q, ¬p ⊢ ¬q | Confuses the conditional with its inverse |

Both are *invalid* — verify with the checker above.
